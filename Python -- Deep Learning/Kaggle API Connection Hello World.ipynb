### Using Command Line

##### In anaconda, you can get to the terminal by going to your environments and pressing the play button dropdown 'open terminal'




##### Go to your terminal and pip install the kaggle package.

> pip install kaggle

##### Select that you accept the kaggle competition rules of the competition by hitting 'enter contest' or late submission'



##### Next, go to your kaggle profile, click edit account; scrolling to the bottom you'll see the Create New API Token



##### A file will be downloaded, now go to a new terminal window run the following command

> mv User/Username/Downloads/kaggle.json User/Username/.kaggle

##### Now your credentials and consent to the competition are confirmed and its time to download the data

> kaggle competitions download -c state-farm-distracted-driver-detection

##### This has the potential of downloading a ton of data, but gives a solid basic overview of how to get your hands on it.
